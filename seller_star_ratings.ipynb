```python
winkel_lijst = [AlgemeneInternationaleBoeken,AlgemeneNederlandstaligeBoeken,AudioenTelevisie,AutoenMotor,BabyHardwaren,
                BabyVerzorging,Beauty,CameraenAccessoires,ComputerAccessoires,ComputerHardware,DierbenodigdhedenenRuitersport,
                Diervoeding,Digitaal,Drank,EducatieveBoeken,Erotiek,FietsKamperenenOutdoorHardwaren,Film,Games,
                GereedschapenVerf,Gezondheid,GrootHuishoudelijk,HobbySpellenenBuitenspeelgoed,Huishouden,KernSpeelgoed,
                KleinHuishoudelijk,KokenenTafelen,Meubelen,MobieleHardwareenWearables,Mode,Muziek,
                PersoonlijkeVerzorgingenHuishoudmiddelen,SanitairenVeiligWonen,Schoenen,SchoolenKantoor,SieradenenHorloges,
                SportHardwaren,SportenOutdoorKledingenSchoenen,TassenReisbagageenModeaccessoires,TelefoonenTabletAccessoires,
                TuinenKerst,Verlichting,WoondecoratieenTextiel]
```

```python
def scorewriter(df):
    with open('ratings_plaza.csv', 'a') as file:
        with open('ratings_plaza.txt', 'a') as txtfile:
            txtfile.write(str(datetime.now()) + '\n')
            file.write('x      ' + str(df.iloc[0][1]) + '\n')
            txtfile.write(str(df.iloc[0][1]) + '\n')
            file.write(str(df.groupby(df['submissionDateTime'].dt.week).mean()['rating']) + '\n')
            txtfile.write(str(df.groupby(df['submissionDateTime'].dt.week).mean()['rating']) + '\n')
            txtfile.write(str(df.groupby(df['submissionDateTime'].dt.week).count()))
            file.write('\n\n')
            txtfile.write('\n\n')
```

```python
for x in winkel_lijst:
    scorewriter(x)
```
